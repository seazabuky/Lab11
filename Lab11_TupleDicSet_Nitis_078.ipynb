การเรียงลำดับตาม keys หรือตามค่า value

```python
#การเรียงลำดับตาม keys หรือตามค่า values
help(sorted)
a={'b': 12, 'a': 18, 'f': 14, 'd': 12, 'e': 18, 'c': 10}
print('--------------------')
print("Ordered by Keys")
for k in sorted(a.keys()):
    print(k,a[k])  
print('--------------------')
print("Ordering Values")
for i in sorted(a.values()):
    print(i)
print('--------------------')
print("Ordered by Values")
for i,j in sorted(a.items(),key=lambda x:x[1]):
    print(i,j)
```

a=['1',3,'a',5,6,'a','2',2,'2',2,14,'m','a','b','k','z',7,13, 3,'1','2']<br/>
โดย key ของ Dictionary จะเป็นไอเทมแต่ละตัวในลิสต์ และ values เป็นจำนวนที่ปรากฎในลิสต์

```python
a=['1',3,'a',5,6,'a','2',2,'2',2,14,'m','a','b','k','z',7,13, 3,'1','2']
result = {}
for x in a:
  count = 0
  for y in a :
    if x == y:
      count += 1
    Dict = {x:count}
  result.update(Dict)
print(result)
```
